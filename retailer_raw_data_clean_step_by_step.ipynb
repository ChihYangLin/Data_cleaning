```python
import pandas as pd
import numpy as np
import os
import random
import math
from datetime import datetime
from datetime import timedelta
```

# 載入資料, 刪除不合理的值, 改欄位名稱

```python
#讀取open data
df = pd.read_csv("OnlineRetail.csv", encoding='utf-8')
df
```

```python
# 改成合適的欄位名稱
rename_dict = {'InvoiceNo':'invoice_no','StockCode':'product_code','Description':'product_name',
              'Quantity':'quantity','InvoiceDate':'date_time','UnitPrice':'unit_price',
               'CustomerID':'customer_id','Country':'region_name'}
df = df.rename(columns = rename_dict)
df
```

# 將customer_id 缺失的位置, 填入數字

```python
# invoice_no欄中有C的列刪掉
df = df.drop(df[df['invoice_no'].apply(lambda x : len(str(x))>6)].index)
# quantity欄中<=0的列刪掉
df = df.drop(df[df['quantity'].apply(lambda x : x <=0)].index)
# unit_price欄中<=0的列刪掉
df = df.drop(df[df['unit_price'].apply(lambda x : x <=0)].index)
df
```

```python
#CustomerID Nan值填上0
df['customer_id'] = df['customer_id'].fillna(0) 
#將值轉成int
df['customer_id'] = df['customer_id'].apply(lambda x : int(x))
```

```python
df[df['customer_id'] == 0]
```

```python
invoice_no_list = list(df[df['customer_id'] == 0]['invoice_no'].drop_duplicates())
num_list = list(range(1,len(invoice_no_list)+1))
invoice_customer_dict = dict(zip(invoice_no_list,num_list))
invoice_customer_dict
```

```python
for i in invoice_customer_dict.keys():
    df['customer_id'][df['invoice_no'] == i] = invoice_customer_dict[i]
```

```python
df
```

#  匯入家樂福資料

```python
# 匯入家樂福資料
df2 = pd.read_csv("Carrefour.csv", encoding='utf-8')
# 去掉空值
df2 = df2.dropna(how='any')
#刪掉df2裡category_id
df2 = df2.drop(columns = ['category_no'])
#刪除ProducName的重複值
df2 = df2.drop_duplicates(subset=['product_name'], ignore_index=True)
df2
```

```python
# 匯入家樂福Top熱銷資料
df3 = pd.read_csv("Carrefour_top.csv", encoding='utf-8')
# 去掉空值
df3 = df3.dropna(how='any')
#刪除df3裡面productname的重複值,剩下熱銷前1277項
df3 = df3.drop_duplicates(subset=['product_name'], ignore_index=True)
df3
```

```python
#將df2裡有df3的值刪掉(df2與df3做差集)
df2 = df2.append(df3)
df2 = df2.append(df3)
df2 = df2.drop_duplicates(subset=['product_name'], keep=False, ignore_index=True)
df2.head(20)
```

```python
#從df裡,將銷售數量總和前幾名的'product_code'放進一個list
top_list = list(df.groupby('product_code').sum().sort_values('quantity', ascending=False).head(len(df3)).index)
#將df3的商品名與 Top項product_code做成字典
top_name_dict = dict(zip(top_list, df3['product_name']))
top_name_dict
```

```python
#將df3的商品價格與 Top項product_code做成字典
top_price_dict = dict(zip(top_list, df3['unit_price']))
top_price_dict
```

```python
#將name代換入對應的product_code裡
for i, j in zip(top_name_dict.keys(),top_name_dict.values()):
    df['product_name'][df['product_code'] == i] = j
#將price代換入對應的product_code裡
for i, j in zip(top_price_dict.keys(),top_price_dict.values()):
    df['unit_price'][df['product_code'] == i] = j
df
```

```python
#建立一個隨機選取數字且不重複的 list
random.seed(123)
random_list = list(range(len(df2)))
random_list = (random.sample(random_list, len(df.groupby('product_code'))-len(df3)))
print(len(random_list))
random_list
```

```python
#建立隨機數字index所對應商品家樂福名稱的 list
product_name_list = list()
for i in random_list:
    product_name_list.append(df2['product_name'][i])
print(len(product_name_list))
product_name_list
```

```python
#建立隨機數字index所對應家樂福價錢的 list
product_price_list = list()
for i in random_list:
    product_price_list.append(df2['unit_price'][i])
print(len(product_price_list))
product_price_list
```

```python
#把product_code放進list
product_code_list = [i[0] for i in df.groupby('product_code')]
print(len(product_code_list))
product_code_list
```

```python
#刪除Top項的product_code
for i in top_list:
    product_code_list.remove(i)
print(len(product_code_list))
product_code_list
```

```python
#將 隨機的product_name 貼上
for i , j in zip(product_code_list, range(len(product_code_list))):
    df['product_name'][df['product_code'] == i] = product_name_list[j]
#將 unit_price 貼上
for i , j in zip(product_code_list,range(len(product_code_list))):
    df['unit_price'][df['product_code'] == i] = product_price_list[j]
df
```

```python
#建立product_category全品項不重複的dataframe
product_category_df = df2.append(df3, ignore_index= True)
#刪除ProductPrice欄位
product_category_df = product_category_df.drop(columns = "unit_price")
product_category_df
```

#  新增 'region_name', 'region_id', 'gender' 欄位

```python
#匯入桃園鄉政區域以及其代號
taoyuan_map = pd.read_csv('taoyuan_region_number.csv', encoding='utf-8')
taoyuan_map.drop(columns = ['Unnamed: 0'], inplace=True)
taoyuan_map
```

```python
#計算customer_id個數
cus_id_num = len(df['customer_id'].value_counts())
print(cus_id_num)
```

```python
# 建立不重複的customer_id list
cus_id_list = list(df['customer_id'].drop_duplicates())

# 建立 region_id 的list
region_no_random_list = random.choices(list(taoyuan_map["region_id"]),k = cus_id_num)

# 建立 gender的 list
sex_list = random.choices(['Male','Female'] , k= cus_id_num )

#建立 customer_region_no_df
customer_region_no_df = pd.DataFrame({'customer_id':cus_id_list,'region_id':region_no_random_list})

#建立 customer_gender_df
customer_gender_df = pd.DataFrame({'customer_id':cus_id_list,'gender':sex_list})

# 與 customer_region_no_df 合併
df = pd.merge(df, customer_region_no_df, on='customer_id')

# 與 customer_gender_df 合併
df = pd.merge(df, customer_gender_df, on='customer_id')
df
```

```python
df = df.drop(columns=['region_name'])
df = pd.merge(df, taoyuan_map, on = 'region_id')
df
```

```python
def process_duplicate_product(df):

    # 建造invoice_no的list, 並去掉重複值
    invoice_no_list = list(df['invoice_no'].drop_duplicates())

    # 把顧客購物清單上, 重複的商品數量做平均, 回傳新的df, 欄位為['product_code','quantity','invoice_no']
    df1 = pd.DataFrame(columns = ['product_code','quantity','invoice_no'])
    for inv_no in invoice_no_list:
        df2 = df[df['invoice_no'] == inv_no].groupby('product_code', as_index =False).agg({'quantity': lambda x : x.sum()})
        df2['invoice_no'] = inv_no
        df1 = df1.append(df2)
        # 觀察進度用
        print((len(df1),inv_no))
    return df1

new_df = process_duplicate_product(df)  
new_df
```

```python
# 刪除quantity欄位
df2 = df.drop(columns=['quantity'])
# 將兩個table做join
df3 = pd.merge(df2,new_df).drop_duplicates(subset=['invoice_no','product_code','customer_id']).reset_index(drop=True)
# 新增subtotal欄位
df3['subtotal'] = df3['unit_price'] * df3['quantity']
# 新增['customer_name']欄位
df3['customer_name'] = 'user_name'
df3
```

```python
# 更改欄位名稱與順序
data = {
    'customer_id':df3['customer_id'],
    'customer_name':df3['customer_name'],
    'gender':df3['gender'],
    'invoice_no':df3['invoice_no'],
    'product_name':df3['product_name'],
    'product_code':df3['product_code'],
    'unit_price':df3['unit_price'],
    'quantity':df3['quantity'],
    'subtotal':df3['subtotal'],
    'region_id':df3['region_id'],
    'region_name':df3['region_name'],
    'category_name':df3['category_name'],
    'date_time':df3['date_time'],   
}
df4 = pd.DataFrame(data)
df4 = df4.sort_values(by = ['date_time','invoice_no'])
df4
```

```python
# 檔案輸出
df4.to_csv('cleaned_retail_data.csv',encoding='utf_8_sig',index=False)
```
